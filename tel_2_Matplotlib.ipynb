```python
"""
一つのエリア内に2種類のグラフ
"""

%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

# グラフ用データの生成
y = np.random.rand(6)* 5

# ラベルや目盛り線の追加
x = ["10", "20", "30", "40", "50"]

d_value = []
for i in [1, 2, 3, 4, 5]:
    d_value.append((10/i) + 10) 

# Dの曲線: 曲線に名前にラベルを付ける
plt.plot(x, d_value, linestyle=":", marker=".", color="g", label="D")

# MCの曲線: 曲線に名前にラベルを付ける
mc_value = []
for i in [1, 2, 4, 6, 7]:
    mc_value.append((i**2)+1)
plt.plot(x, mc_value, linestyle="--", marker="x", color="blue", label="MC")


# 全体の設定
plt.title("sample")
plt.xlabel("Q")
plt.ylabel("P")
plt.legend(loc="lower right")
plt.grid(True)
```

```python
"""
2つのグラフを別々に表示
"""

%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

# グラフ用データの生成
y = np.random.rand(6)* 5

# ラベルや目盛り線の追加
x = ["10", "20", "30", "40", "50"]


# 全体の描画領域を生成
fig = plt.figure()

# 1つのグラフ用の描画領域の生成
# 描画領域　fig を 1 x 2 の領域とし、1つ目として plot1 を配置する。 
plot1 = fig.add_subplot(1, 2, 1)

d_value = []
for i in [1, 2, 3, 4, 5]:
    d_value.append((10/i) + 10) 

# Dの曲線: 曲線に名前にラベルを付ける
plot1.plot(x, d_value, linestyle=":", marker=".", color="g", label="D")

# ----

# 2つ目のグラフ用の描画領域の生成
# 描画領域　fig を 1 x 2 の領域とし、2つ目として plot1 を配置する。
plot2 = fig.add_subplot(1, 2, 2)

# MCの曲線: 曲線に名前にラベルを付ける
mc_value = []
for i in [1, 2, 4, 6, 7]:
    mc_value.append((i**2)+1)
plot2.plot(x, mc_value, linestyle="--", marker="x", color="blue", label="MC")

# 全体の設定
plt.title("sample")
plt.xlabel("production quantity")
plt.ylabel("price")
plt.legend(loc="lower right")
plt.grid(True)
```

```python
"""
近似値線
"""

%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

# グラフ用データの生成
y = np.random.rand(5)*2
#y = [11, 25, 29, 42, 55]

# x の値
x = [1, 2, 3, 4, 5]

# 1つのグラフ用の描画領域の生成
# 描画領域　fig を 1 x 2 の領域とし、1つ目として plot1 を配置する。 
plt.scatter(x,y)

# 近似直線のタプルを取得
# p = polyfit(x,y,n) は、y のデータに対して 
# (最小二乗的に) 最適な近似となる n 次の多項式 p(x) の係数を返す
p = np.polyfit(x, y, 2)
print(f"polyfit {p} - {type(p)}")

    
# # 近似値を表す関数を生成する
f = np.poly1d(p)
print(f"poly1dで生成された関数: {f}")
plt.plot(x, f(x))


# 全体の設定
plt.title("sample")
plt.xlabel("Q")
plt.ylabel("P")
#plt.legend(loc="lower right")
plt.grid(True)
```
