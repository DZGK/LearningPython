```python
# イメージファイルをモノクロに変換して表示する

from PIL import Image
import matplotlib.pyplot as plt

# イメージファイルの読み込み
img = Image.open("sample.png")
print(type(img))
# イメージファイルを Jupyter Lab 上に表示する。
plt.imshow(img)

# モノクロが画像への変換
# Matplotlibのデフォルトのカラーマップ設定では、暗いところから明るくなるにつれて 青 → 緑 → 黄色 と色付けされる。
img2 = img.convert("L")
plt.imshow(img2)

# 加工後のファイルを保存する。保存されたファイルはモノクロ(白黒)で保存される点に注意。
img2.save("sample2.png")
```

```python
# イメージファイルをモノクロに変換して表示する

from PIL import Image
import matplotlib.pyplot as plt

# イメージファイルの読み込み
img = Image.open("sample.png")
print(type(img))
# イメージファイルを Jupyter Lab 上に表示する。
plt.imshow(img)

# モノクロが画像への変換
# 白黒で表示する場合、 cmap="gray" カラーマップを指定する必要がある。
img2 = img.convert("L")
plt.imshow(img2, cmap="gray")
```

```python
# 画像のサイズの変換
from PIL import Image
import matplotlib.pyplot as plt

img = Image.open("sample.png")
print(img.size)
plt.imshow(img)

img2 = img.resize((250, 375))
plt.imshow(img2)
```

```python
# 画像を回転する
from PIL import Image
import matplotlib.pyplot as plt

img = Image.open("sample.png")
img_rotated = img.rotate(45, expand=True)
plt.imshow(img_rotated)

```

```python
# 画像の平行移動
from PIL import Image
import matplotlib.pyplot as plt

img = Image.open("sample.png")
img_translated = img.rotate(0, translate=(200, 100), expand=True)
plt.imshow(img_translated)

```

```python
# 画像データを NumPy 配列に変換する方法
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

img = Image.open("sample.png")
img_ndarray = np.array(img)
print(type(img_ndarray))

# ndarrayには、画像の左上から右下に向けて、1ピクセルずつ、RGB（赤、緑、青）の色の強さが0から255までの256段階の値として格納される。
## 画像の縦の長さ, 横の長さ の2次元配列。3つ目の "3" は 2次元配列を3個持っているという意味。この 3つは、[0] = R, [1] = G, [2] = B を表す。
print(img_ndarray.shape)
# print(img_ndarray)

## 座標 (0,0) つまり一番左上の色が　R:116, G: 166, B: 218 であることを表す。
print(img_ndarray[0, 0, :])


# モノクロ画像の ndarray への変換
img_gray = img.convert("L")
img_gray_ndarray = np.array(img_gray)
print(f"gray: {img_gray_ndarray.shape}")
```

```python
# 数値化した画像データの利用
# ヒストグラム (色の明るさの分布) の作成
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

img = Image.open("sample.png")
img_ndarray = np.array(img)

hist_c, binsc = np.histogram(img_ndarray.flatten(), bins=256)
plt.plot(hist_c)
```

```python
# 数値化した画像データの利用
# 赤色に対するヒストグラム (色の明るさの分布) の作成
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

img = Image.open("sample.png")
img_ndarray = np.array(img)

hist_c, binsc = np.histogram(img_ndarray[:, :, 0].flatten(), bins=256)
plt.plot(hist_c)
```

```python
# ndarray データの画像への変換: その1: imshow() を使う

import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

img = Image.open("sample.png")
img_ndarray = np.array(img)
img_red = img_ndarray[:, :, 0]
plt.imshow(img_red)
```

```python
# ndarray データの画像への変換: その2: Image.fromarray() を使う

import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

img = Image.open("sample.png")
img_ndarray = np.array(img)
img_green = img_ndarray[:, :, 1]

img2 = Image.fromarray(np.uint8(img_green))
plt.imshow(img2)
```
