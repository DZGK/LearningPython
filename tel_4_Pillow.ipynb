```python
# イメージファイルをモノクロに変換して表示する

from PIL import Image
import matplotlib.pyplot as plt

# イメージファイルの読み込み
img = Image.open("sample.png")
print(type(img))
# イメージファイルを Jupyter Lab 上に表示する。
plt.imshow(img)

# モノクロが画像への変換
# Matplotlibのデフォルトのカラーマップ設定では、暗いところから明るくなるにつれて 青 → 緑 → 黄色 と色付けされる。
img2 = img.convert("L")
plt.imshow(img2)

# 加工後のファイルを保存する。保存されたファイルはモノクロ(白黒)で保存される点に注意。
img2.save("sample2.png")
```

```python
# イメージファイルをモノクロに変換して表示する

from PIL import Image
import matplotlib.pyplot as plt

# イメージファイルの読み込み
img = Image.open("sample.png")
print(type(img))
# イメージファイルを Jupyter Lab 上に表示する。
plt.imshow(img)

# モノクロが画像への変換
# 白黒で表示する場合、 cmap="gray" カラーマップを指定する必要がある。
img2 = img.convert("L")
plt.imshow(img2, cmap="gray")
```
